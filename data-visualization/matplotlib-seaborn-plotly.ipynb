## Trac basique  <a class="anchor" id="section_2_2"></a>

La fonction ```plot``` permet de tracer un trac basique. Elle prend en argument une squences de donnes pour les abscices (```X```) et une squence de donnes pour les ordonnes (```Y```).

````python
import numpy as np
import pandas as pd
from math import exp

# cration d'une squence de d'abscices
X = np.linspace(1, 10, num=100)

# cration d'une squence en ordonnes
Y = [exp(elt) for elt in X]
````

````python
# tracer X en fonction de Y
plt.plot(X,Y)
# affichage du graphe
plt.show()
````

* <span class="text-warning"> $\textbf{Remarque}$: Si on ne travaille pas sur un Jupyter notebook, il faut ajouter la commande ```plt.show()```pour afficher la figure </span> 

### Edition de paramtres  <a class="anchor" id="section_2_2_1"></a>

* <span class="text-muted"> Ajouter le titre "Fonction exponentielle" </span> 
* <span class="text-muted"> Ajouter le titre "Axe X" pour l'axe des abscices  </span> 
* <span class="text-muted"> Ajouter le titre "Axe Y" pour l'axe des ordonnes  </span> 
* <span class="text-muted"> Dlimiter l'affichage graphique pour $X \in [8,10]$ et $Y \in [5000,15 000]$   </span> 

````python
# tracer de la fonction
plt.plot(X,Y) 
# ajouter un titre
plt.title('Fonction exponentielle')
# ajouter un label  l'axe des abscices
plt.xlabel('Axe X')
# ajouter un label  l'axe des ordonnes
plt.ylabel('Axe Y')
# dlimiter l'affichage sur l'abscices
plt.xlim(8,10)
# dlimiter l'affichage sur les ordonnes
plt.ylim(5000, 15000)
# affichage du graphe
plt.show()
````

* <span class="text-muted"> Tracer la courbe en pointills, orange (```#FF8C00```) </span>


````python
# tracer de la courbe
plt.plot(X, Y, color="#FF8C00", ls=':') 
# ajout du titre
plt.title('Fonction exponentielle')
# ajouter un label  l'axe des abscices
plt.xlabel('Axe X')
# ajouter un label  l'axe des ordonnes
plt.ylabel('Axe Y')
# affichage du graphe
plt.show()
````

* <span class="text-muted"> Crer plusieurs fentres grce  subplot() </span>

````python
# dlimitation des subplots
fig, axes = plt.subplots(nrows=2, ncols=2)
# tracer des 4 figures
axes[0][0].plot(X, Y)
axes[0][1].plot(X, Y, color="red", ls=':')
axes[1][0].plot(X, Y, color="green", ls='--')
axes[1][1].plot(X, Y, color="purple", ls='-.')
````

# Plotly

Plotly dont la documentation se trouve [ici](https://plotly.com/python), permet de raliser des visualisations graphiques interactives.

* <span class="text-muted"> Importer la bibliothque ```plotly``` sous le nom ```px``` </span>

````python
import plotly.express as px
````

* <span class="text-muted"> Importation du data set ```gapminder```</span>

````python
df5 = px.data.gapminder()
df5.tail()
````

* Trac de la population par continent au fil des annes

````python
px.bar(df5, x="continent", y="pop", color="continent",
      animation_frame="year", animation_group="country", range_y=[0,4000000000])
````

* Trac de l'esprance de vie en fonction du PIB par pays au fil des annes

````python
px.scatter(df5, x="gdpPercap", y="lifeExp", animation_frame="year", animation_group="country",
           size="pop", color="continent", hover_name="country",
           log_x=True, size_max=55, range_x=[100,100000], range_y=[25,90])
````
